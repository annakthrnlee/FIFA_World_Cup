```python
#  Import and read the dataset
import pandas as pd 
Revised_df = pd.read_csv("international_matches.csv")
Revised_df.head()
```

```python
# Check the datatypes of your columns
Revised_df.dtypes
```

```python
# Drop unecessary columns
Revised_df.drop(["shoot_out", "home_team_goalkeeper_score", "away_team_goalkeeper_score", "home_team_mean_defense_score", "home_team_mean_offense_score", "home_team_mean_midfield_score"], axis=1, inplace=True)
```

```python
# Drop unecessary columns
Revised_df.drop(["away_team_mean_defense_score", "away_team_mean_offense_score", "away_team_mean_midfield_score", "home_team_score", "away_team_score"], axis=1, inplace=True)
Revised_df
```

```python
# Drop unecessary columns
Revised_df.drop(["city", "neutral_location", "home_team_total_fifa_points", "away_team_total_fifa_points"], axis=1, inplace=True)
Revised_df
```

```python
# Since the dates go back to 1930, I'm removing all dates up to 2000
Revised_dates_df = Revised_df[(Revised_df['date'] >= '2000-01-01')
                          & (Revised_df['date'] < '2022-08-22')]

Revised_dates_df
```

```python
# Drop all null values 'columns' from dataset
Revised_dates_df = Revised_dates_df.dropna(axis=1)
Revised_dates_df
```

```python
# Sort home ranking by ascending order 
Ranking_home_df = Revised_dates_df.sort_values(by='home_team_fifa_rank', ascending=True)
Ranking_home_df
```

```python
# Sort away ranking by ascending order
Ranking_away_df = Revised_dates_df.sort_values(by='away_team_fifa_rank', ascending=True)
Ranking_away_df
```
