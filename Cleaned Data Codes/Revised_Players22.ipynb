```python
#  Import and read the dataset
import pandas as pd 
Revised_df = pd.read_csv("players_22.csv")
Revised_df.head()
```

```python
# Check the datatypes of your columns
Revised_df.dtypes
```

```python
Revised_df.drop(["player_url", "club_team_id", "club_name", "club_jersey_number"], axis=1, inplace=True)
```

```python
Revised_df.drop(["club_loaned_from", "club_joined", "club_contract_valid_until", "player_positions", "overall"], axis=1, inplace=True)
```

```python
Revised_df.drop(["nation_position", "nation_jersey_number", "weak_foot", "skill_moves", "international_reputation", "work_rate"], axis=1, inplace=True)
```

```python
Revised_df.drop(["release_clause_eur", "player_tags", "pace", "shooting", "pass", "dribbling", "defending", "physic", "attacking_crossing"], axis=1, inplace=True)
```

```python
Revised_df.drop(["attacking_finishing", "attacking_heading_accuracy", "attacking_short_passing", "attacking_volleys", "skill_dribbling", "skill_curve"], axis=1, inplace=True)
```

```python
Revised_df.drop(["skill_fk_accuracy", "skill_long_passing", "skill_ball_control", "movement_acceleration", "movement_sprint_speed", "movement_agility", "movement_reactions"], axis=1, inplace=True)
```

```python
Revised_df.drop(["movement_balance", "power_shot_power", "power_jumping", "power_stamina", "power_strength", "power_long_shots", "mentality_aggression", "mentality_interceptions", "mentality_positioning", "mentality_vision"], axis=1, inplace=True)
```

```python
Revised_df.drop(["mentality_penalties", "mentality_composure", "potential", "value_eur", "wage_eur", "defending_marking_awareness", "defending_standing_tackle", "defending_sliding_tackle"], axis=1, inplace=True)
```

```python
Revised_df.drop(["goalkeeping_diving", "goalkeeping_handling", "goalkeeping_kicking", "goalkeeping_positioning", "goalkeeping_reflexes", "goalkeeping_speed", "club_position", "nationality_id", "league_name", "league_level"], axis=1, inplace=True)
```

```python
# Yes, lots and lots and lots of columns had to be dropped. Our goal is to compare highest ranked teams 2022 with players 2022 age, height, weight, body_type, and nationality
```

```python
# Drop all players who don't play in the necessary leagues to match our international_matches dataframe 
nan_value = float("NaN")
Revised_df.replace("", nan_value, inplace=True)
Revised_df.dropna(subset = ["nation_team_id"], inplace=True)
Revised_df
```
